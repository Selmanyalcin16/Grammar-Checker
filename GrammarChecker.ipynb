```python
from transformers import pipeline

pipe = pipeline("text2text-generation", model="grammarly/coedit-large")
```

```python
pipe("I has go to that mall since ı were a kid")
```

```python
def ner(text):
  output=pipe(text)
  return output[0]
```

```python
import gradio as gr
```

```python
def ner(text):
  output=pipe(text)
  return output[0]
```

```python
app_inputs = gr.Textbox(lines=2, placeholder="Enter sentence here...")
```

```python
interface = gr.Interface(fn=ner,
                        inputs=app_inputs,
                         outputs='text',
                        title='Grammar Checker')
```

```python
interface.launch()
```
